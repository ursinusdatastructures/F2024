# The Fast Ones

```python
def log2(N):
    res = 0
    i = N
    while i > 1:
        i = i // 2
        res += 1
    return res


def logstar(N):
    res = 0
    i = N
    while i > 1:
        i = log2(i)
        res += 1
    return res

print(log2(1023))
```

```python
# NOTE: Up to 2^410 atoms in the universe
# log2(2^410) = 410
# log2(log2(2^410)) = 8
# logstar(2^410) <= 5

import matplotlib.pyplot as plt
import numpy as np

N = 2**np.arange(0, 60)
plt.plot(N, np.ones(N.size))
plt.plot(N, [logstar(Ni) for Ni in N])
plt.plot(N, [log2(Ni) for Ni in N])
plt.plot(N, [log2(log2(Ni)) for Ni in N])
plt.legend(["$1$", "log*(N)", "log(N)", "log(log(N))"])
plt.gca().set_xscale('log')
plt.xlabel("N")
plt.title("The Fast Ones");
```

# The OK Ones

```python
N = np.arange(1, 100)
plt.plot(N, N)
plt.plot(N, N*np.log2(N))
plt.plot(N, N*np.log2(N)*np.log2(N))
plt.plot(N, N*N)
plt.plot(N, np.sqrt(N))
plt.legend(["N", "N(logN)", "N(logN)(logN)", "$N^2$", "$\sqrt{N}$"])
plt.xlabel("N")
plt.title("The OK Ones");
```

## The Slow Ones

```python
def fac(N):
    res = 1 
    if N > 1:
        res = N*fac(N-1)
    return res
    
N = np.arange(1, 20)
plt.plot(2**N)
plt.plot([fac(Ni) for Ni in N])
plt.gca().set_yscale('log')
plt.legend(["$2^N$", "$N!$"])
plt.xlabel("N")
plt.title("The Slow Ones")
```
