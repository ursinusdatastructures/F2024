```python
## Ex) Loop through all triples in an array
def count_triples(N, verbose=False):
    """
    N: int
        The number of elements in my set
    verbose: bool
        True if I want to print out the triples in zero-indexed
        representation of the set [0, 1, 2, ..., N-1]
    """
    n_triples = 0
    for i in range(N):
        for j in range(i+1, N):
            for k in range(j+1, N):
                if verbose:
                    print(i, j, k)
                n_triples += 1
    return n_triples

Ns = np.arange(1, 45)
triples = [count_triples(N) for N in Ns]
plt.plot(Ns, triples)
plt.plot(Ns, Ns**3/8)
plt.plot(Ns, 4*Ns**2)
plt.plot(Ns, Ns**4/300)
plt.legend(["Triples", "$N^3/8$", "$4 N^2$", "$N^4/300$"])
plt.xlabel("N")
plt.ylabel("Num triples")
```

## An algorithm that is O(log(N))
### And algorithms that are O(log(log(N)) and O(log*(N))

```python
def log_prototype(N):
    ## How many times I have to halve a number to get to 1
    counter = 0
    while N > 1:
        N = N//2
        counter += 1
    return counter

def loglog_prototype(N):
    ## How many times I have to take the square root
    ## of a number to get to 1
    counter = 0
    while N > 1:
        N = np.floor(N**0.5) # sqrt(N)
        counter += 1
    return counter

def logstar_prototype(N):
    ## How many times I have to take the log to get to 1
    counter = 0
    while N > 1:
        N = np.floor(np.log(N))
        counter += 1
    return counter



Ns = np.arange(2, 10000000000, 10000000)
plt.plot(Ns, [log_prototype(N) for N in Ns])
plt.plot(Ns, [loglog_prototype(N) for N in Ns])
plt.plot(Ns, [logstar_prototype(N) for N in Ns])
plt.legend(["$\\log(N)$", "$\\log(\\log(N))$", "$\\log*(N)$"])
```
